```python
# =============================================================================
# SKRIPSI: PREDIKSI BANJIR JAKARTA SELATAN (FINAL FIXED & ENHANCED)
# Fitur Baru: Visualisasi TP/FP/FN pada Time Series
# =============================================================================

import os
import ee
import requests
import joblib
import optuna
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
import matplotlib.dates as mdates

from optuna.integration import TFKerasPruningCallback
from sklearn.preprocessing import QuantileTransformer
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score, recall_score, f1_score, roc_curve, precision_recall_curve, auc
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Bidirectional
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# 1. SETUP LINGKUNGAN
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 11})

# Folder Simpan
SAVE_DIR = 'laporan_skripsi_final_fix'
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)
print(f"ðŸ“‚ Folder '{SAVE_DIR}' siap.")

# =============================================================================
# 2. DATA LOAD & CLEANING (BAB 3)
# =============================================================================
MY_PROJECT = 'plenary-vim-481714-s8'  # <--- GANTI ID PROJECT ANDA

print("\n" + "="*50)
print("TAG 1: DATA ACQUISITION & INSPECTION")
print("="*50)

try:
    ee.Initialize(project=MY_PROJECT)
except:
    ee.Authenticate()
    ee.Initialize(project=MY_PROJECT)

lat = -6.3120; lon = 106.8180
point = ee.Geometry.Point([lon, lat])
start_date = '2001-01-01'; end_date = '2024-01-01'

# 2.1 Load CHIRPS
chirps = ee.ImageCollection("UCSB-CHG/CHIRPS/DAILY").filterDate(start_date, end_date)
data_raw = chirps.getRegion(point, scale=5500).getInfo()
df_chirps = pd.DataFrame(data_raw[1:], columns=data_raw[0])
df_chirps['date'] = pd.to_datetime(df_chirps['time'], unit='ms')
df_chirps = df_chirps.set_index('date').sort_index()
df_chirps['Hujan'] = pd.to_numeric(df_chirps['precipitation']).fillna(0)

# 2.2 Load NASA POWER
str_start = start_date.replace('-', ''); str_end = end_date.replace('-', '')
base_url = "https://power.larc.nasa.gov/api/temporal/daily/point"
params = {'parameters': 'T2M,RH2M,WS2M,PS,T2MDEW', 'community': 'RE',
          'longitude': lon, 'latitude': lat, 'start': str_start, 'end': str_end, 'format': 'JSON'}
response = requests.get(base_url, params=params)
data_nasa = response.json()
feat = data_nasa['properties']['parameter']
df_nasa = pd.DataFrame([feat['T2M'], feat['RH2M'], feat['WS2M'], feat['PS'], feat['T2MDEW']]).T
df_nasa.columns = ['Suhu', 'Kelembaban', 'Angin', 'Tekanan', 'TitikEmbun']
df_nasa.index = pd.to_datetime(df_nasa.index, format='%Y%m%d')

# Gabung Data
df_raw = pd.concat([df_chirps[['Hujan']], df_nasa], axis=1, join='inner')

# PRINT INFO DATASET
print("\n[INFO DATASET AWAL]")
print(df_raw.info())
print("\n[STATISTIK DESKRIPTIF]")
print(df_raw.describe().T)

# Cleaning
df = df_raw.fillna(method='ffill').fillna(method='bfill')

# VISUALISASI 1: Time Series Hujan
plt.figure(figsize=(12, 4))
plt.plot(df.index, df['Hujan'], label='Curah Hujan', color='blue', alpha=0.6)
plt.title('Time Series Data Curah Hujan (2001-2024)')
plt.ylabel('Hujan (mm)')
plt.legend()
plt.savefig(f'{SAVE_DIR}/1_timeseries_hujan_raw.png')
plt.show()

# =============================================================================
# 3. FEATURE ENGINEERING & EDA (BAB 3)
# =============================================================================
print("\n" + "="*50)
print("TAG 2: FEATURE ENGINEERING & EDA")
print("="*50)

df['Status'] = df['Hujan'].apply(lambda x: 1 if x > 20 else 0)
df['Dew_Depression'] = df['Suhu'] - df['TitikEmbun']
df['Delta_PS'] = df['Tekanan'].diff()
df['Delta_RH'] = df['Kelembaban'].diff()
df['Mean_7d'] = df['Hujan'].shift(1).rolling(7).mean()
df['Std_7d'] = df['Hujan'].shift(1).rolling(7).std()
df['Lag1'] = df['Hujan'].shift(1)
df['Lag2'] = df['Hujan'].shift(2)
df['Sin_Time'] = np.sin(df.index.map(pd.Timestamp.timestamp) * (2 * np.pi / (365.25*24*3600)))

df = df.dropna()
features = ['Suhu', 'Kelembaban', 'Angin', 'Tekanan', 'Dew_Depression', 'Delta_PS', 'Delta_RH',
            'Mean_7d', 'Std_7d', 'Lag1', 'Lag2', 'Sin_Time']
X_eda = df[features]
y_eda = df['Status']

# VISUALISASI 2: Pie Chart Rasio
plt.figure(figsize=(6, 6))
counts = y_eda.value_counts()
plt.pie(counts, labels=[f'Aman ({counts[0]})', f'Banjir ({counts[1]})'], autopct='%1.1f%%',
        colors=['#AED6F1', '#E74C3C'], startangle=90, explode=(0, 0.1))
plt.title('Rasio Dataset Awal (Imbalanced)')
plt.savefig(f'{SAVE_DIR}/2_eda_ratio.png')
plt.show()

# VISUALISASI 3: Heatmap Korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(df[features + ['Status']].corr(), cmap='coolwarm', annot=False)
plt.title('Matriks Korelasi Fitur')
plt.savefig(f'{SAVE_DIR}/3_eda_heatmap.png')
plt.show()

# VISUALISASI 4: Boxplot
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.boxplot(x='Status', y='Lag1', data=df, palette=['skyblue', 'salmon'])
plt.title('Hujan Kemarin vs Banjir')
plt.subplot(1, 2, 2)
sns.boxplot(x='Status', y='Dew_Depression', data=df, palette=['skyblue', 'salmon'])
plt.title('Dew Depression vs Banjir')
plt.savefig(f'{SAVE_DIR}/4_eda_boxplot.png')
plt.show()

# =============================================================================
# 4. PREPROCESSING & SMOTE (BAB 3)
# =============================================================================
print("\n" + "="*50)
print("TAG 3: PREPROCESSING & SMOTE")
print("="*50)

# Normalisasi
scaler = QuantileTransformer(output_distribution='normal', random_state=SEED)
X_scaled = scaler.fit_transform(X_eda)

# Windowing
LOOK_BACK = 14
def create_dataset(X, y, look_back):
    Xs, ys = [], []
    for i in range(len(X) - look_back):
        Xs.append(X[i:(i + look_back)])
        ys.append(y.iloc[i + look_back])
    return np.array(Xs), np.array(ys)

X_seq, y_seq = create_dataset(X_scaled, y_eda, LOOK_BACK)

# Split Chronological
split_idx = int(len(X_seq) * 0.8)
X_train_raw, X_test = X_seq[:split_idx], X_seq[split_idx:]
y_train_raw, y_test = y_seq[:split_idx], y_seq[split_idx:]

# SMOTE
n_train, t_steps, n_feats = X_train_raw.shape
X_train_flat = X_train_raw.reshape(n_train, t_steps * n_feats)
smote = SMOTE(random_state=SEED)
X_train_smote, y_train_smote = smote.fit_resample(X_train_flat, y_train_raw)
X_train = X_train_smote.reshape(-1, t_steps, n_feats)
y_train = y_train_smote

# VISUALISASI 5: SMOTE Comparison
plt.figure(figsize=(8, 5))
labels = ['Aman', 'Banjir']
x = np.arange(len(labels))
width = 0.35
plt.bar(x - width/2, np.bincount(y_train_raw), width, label='Sebelum SMOTE', color='skyblue')
plt.bar(x + width/2, np.bincount(y_train_smote), width, label='Sesudah SMOTE', color='salmon')
plt.xticks(x, labels); plt.legend()
plt.title('Dampak SMOTE pada Data Training')
plt.savefig(f'{SAVE_DIR}/5_smote_comparison.png')
plt.show()

# =============================================================================
# 5. MODELING (OPTUNA + ENSEMBLE) (BAB 3 & 4)
# =============================================================================
print("\n" + "="*50)
print("TAG 4: HYPERPARAMETER TUNING & TRAINING")
print("="*50)

def objective(trial):
    units = trial.suggest_categorical('units', [32, 64, 128])
    dropout = trial.suggest_float('dropout', 0.2, 0.5)
    lr = trial.suggest_float('lr', 1e-4, 1e-2, log=True)

    model = Sequential()
    model.add(Bidirectional(LSTM(units, return_sequences=False), input_shape=(LOOK_BACK, n_feats)))
    model.add(Dropout(dropout))
    model.add(BatchNormalization())
    model.add(Dense(16, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # Paksa nama metric 'auc'
    model.compile(optimizer=Adam(learning_rate=lr), loss='binary_crossentropy', metrics=[tf.keras.metrics.AUC(name='auc')])

    history = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                        epochs=8, batch_size=64, verbose=0,
                        callbacks=[TFKerasPruningCallback(trial, 'val_auc')])
    return max(history.history['val_auc'])

# OPTUNA TUNING
print("... Sedang mencari parameter terbaik ...")
study = optuna.create_study(direction='maximize', pruner=optuna.pruners.MedianPruner())
study.optimize(objective, n_trials=10)
best_params = study.best_params
print(f"âœ¨ Parameter Terbaik: {best_params}")

# Training Ensemble
models = []
history_list = []

for i in range(3):
    print(f"Melatih Model Member {i+1}...")
    tf.keras.backend.clear_session()
    model = Sequential()
    model.add(Bidirectional(LSTM(best_params['units'], return_sequences=False), input_shape=(LOOK_BACK, n_feats)))
    model.add(Dropout(best_params['dropout']))
    model.add(BatchNormalization())

    if i == 1: model.add(Dense(32, activation='relu'))
    else: model.add(Dense(16, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer=Adam(learning_rate=best_params['lr']), loss='binary_crossentropy', metrics=[tf.keras.metrics.AUC(name='auc')])

    es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    hist = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                     epochs=40, batch_size=64, verbose=0, callbacks=[es], shuffle=True)
    models.append(model)
    history_list.append(hist)

# VISUALISASI 6: Training History
# Mencari key yang benar (auc/auc_1/val_auc)
hist_keys = history_list[0].history.keys()
auc_key = [k for k in hist_keys if 'auc' in k and 'val' not in k][0]
val_auc_key = [k for k in hist_keys if 'auc' in k and 'val' in k][0]

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_list[0].history['loss'], label='Train Loss')
plt.plot(history_list[0].history['val_loss'], label='Val Loss')
plt.title('Kurva Loss (Model 1)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_list[0].history[auc_key], label='Train AUC')
plt.plot(history_list[0].history[val_auc_key], label='Val AUC')
plt.title('Kurva AUC (Model 1)')
plt.legend()
plt.savefig(f'{SAVE_DIR}/6_training_history.png')
plt.show()

# =============================================================================
# 6. EVALUASI LENGKAP (BAB 4)
# =============================================================================
print("\n" + "="*50)
print("TAG 6: EVALUASI & THRESHOLD CALIBRATION")
print("="*50)

yhats = [m.predict(X_test, verbose=0) for m in models]
y_pred_avg = np.mean(yhats, axis=0)

# AUTO-CALIBRATION
candidates = []
print("Scanning Threshold Terbaik...")
for t in np.arange(0.10, 0.40, 0.01):
    y_temp = (y_pred_avg > t).astype(int)
    rec = recall_score(y_test, y_temp)
    acc = accuracy_score(y_test, y_temp)
    # Filter: Recall minimal 50%
    if rec >= 0.50:
        candidates.append({'t': t, 'rec': rec, 'acc': acc})

if len(candidates) > 0:
    # Ambil akurasi tertinggi dari kandidat yang lolos recall
    best_c = max(candidates, key=lambda x: x['acc'])
    FINAL_THRESH = best_c['t']
else:
    FINAL_THRESH = 0.15

print(f"ðŸ† Threshold Terpilih Otomatis: {FINAL_THRESH:.2f}")

y_final = (y_pred_avg > FINAL_THRESH).astype(int)

# 1. Classification Report
print("\n[CLASSIFICATION REPORT FINAL]")
print(classification_report(y_test, y_final, target_names=['AMAN', 'BANJIR']))

# VISUALISASI 7: Confusion Matrix
cm = confusion_matrix(y_test, y_final)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', xticklabels=['Aman', 'Banjir'], yticklabels=['Aman', 'Banjir'])
plt.title(f'Confusion Matrix (Threshold {FINAL_THRESH:.2f})')
plt.savefig(f'{SAVE_DIR}/7_confusion_matrix.png')
plt.show()

# VISUALISASI 8: ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_avg)
roc_auc_val = auc(fpr, tpr)
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc_val:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.savefig(f'{SAVE_DIR}/8_roc_curve.png')
plt.show()

# =============================================================================
# 7. FEATURE IMPORTANCE & ACTUAL VS PREDICTED (BAB 4)
# =============================================================================
print("\n" + "="*50)
print("TAG 7: FEATURE IMPORTANCE & PREDICTION ANALYSIS")
print("="*50)

# Feature Importance
baseline_auc = roc_auc_score(y_test, y_pred_avg)
importances = []
for i, col in enumerate(X_eda.columns):
    X_perm = X_test.copy()
    for t in range(LOOK_BACK): np.random.shuffle(X_perm[:, t, i])

    yhats_perm = [m.predict(X_perm, verbose=0) for m in models]
    y_pred_perm = np.mean(yhats_perm, axis=0)
    drop = baseline_auc - roc_auc_score(y_test, y_pred_perm)
    importances.append({'Feature': col, 'Importance': max(drop, 0)})

df_imp = pd.DataFrame(importances).sort_values(by='Importance', ascending=False)

# VISUALISASI 9: Feature Importance
plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=df_imp.head(15), palette='viridis')
plt.title('Top 15 Fitur Paling Berpengaruh')
plt.tight_layout()
plt.savefig(f'{SAVE_DIR}/9_feature_importance.png')
plt.show()

# =============================================================================
# [FIXED & ENHANCED] VISUALISASI 10: ACTUAL VS PREDICTED (DIAGNOSTIC PLOT)
# =============================================================================
# Perbaikan: Menggunakan .flatten() agar tidak error dimensi
# Fitur: Mewarnai titik berdasarkan TP, FP, FN, TN

zoom_days = 200 # Lihat 200 hari terakhir
dates_zoom = df.index[split_idx + LOOK_BACK:][-zoom_days:]

# Flatten array agar dimensi aman (N,)
y_true_zoom = y_test[-zoom_days:].flatten()
y_pred_zoom = y_final[-zoom_days:].flatten()

# Buat Masking untuk pewarnaan
# TP: Prediksi 1, Asli 1 (Success)
mask_tp = (y_true_zoom == 1) & (y_pred_zoom == 1)
# FP: Prediksi 1, Asli 0 (False Alarm)
mask_fp = (y_true_zoom == 0) & (y_pred_zoom == 1)
# FN: Prediksi 0, Asli 1 (Missed / Bahaya)
mask_fn = (y_true_zoom == 1) & (y_pred_zoom == 0)

plt.figure(figsize=(15, 6))

# 1. Plot Garis Ground Truth (Garis Putus-putus Hitam)
plt.plot(dates_zoom, y_true_zoom, label='Aktual (Ground Truth)', color='black', alpha=0.3, linestyle='--')

# 2. Plot Area Prediksi (Area Merah Tipis)
plt.fill_between(dates_zoom, 0, y_pred_zoom, color='red', alpha=0.1, label='Sinyal Prediksi Model')

# 3. Plot Titik-Titik Penting (TP, FP, FN)
plt.scatter(dates_zoom[mask_tp], y_true_zoom[mask_tp], color='green', s=100, marker='*', label='TP (Sukses Deteksi)')
plt.scatter(dates_zoom[mask_fp], y_pred_zoom[mask_fp], color='red', s=50, marker='X', label='FP (Alarm Palsu)')
plt.scatter(dates_zoom[mask_fn], y_true_zoom[mask_fn], color='orange', s=80, marker='^', label='FN (Gagal Deteksi)')

plt.title(f'Analisis Diagnostik Prediksi Banjir ({zoom_days} Hari Terakhir)\nThreshold: {FINAL_THRESH:.2f}')
plt.yticks([0, 1], ['Aman', 'Banjir'])
plt.legend(loc='upper left', bbox_to_anchor=(0, -0.1), ncol=4)
plt.tight_layout()
plt.savefig(f'{SAVE_DIR}/10_diagnostic_plot_tp_fp.png')
plt.show() # <--- TAMPILKAN

print("\nâœ… SELESAI. Semua output telah ditampilkan dan disimpan.")
```

```python
# =============================================================================
# SKRIPSI FINAL: REGRESI LSTM (ROBUST HUBER LOSS) - LOKASI RAGUNAN
# Strategi: Mempertahankan Regresi dengan Arsitektur Tahan Banting
# =============================================================================

import ee
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Setup
tf.random.set_seed(42)
np.random.seed(42)
sns.set_style("whitegrid")

# =============================================================================
# 1. AMBIL DATA HYBRID (CHIRPS + NASA)
# =============================================================================
# --- PENTING: ISI PROJECT ID ANDA DISINI ---
MY_PROJECT = 'plenary-vim-481714-s8'

print("ðŸš€ [1/5] Download Data Hybrid (Lokasi: Ragunan)...")
try:
    ee.Initialize(project=MY_PROJECT)
except:
    ee.Authenticate()
    ee.Initialize(project=MY_PROJECT)

# KOORDINAT RAGUNAN, JAKARTA SELATAN
lat = -6.3120  # Koordinat sekitar Kebun Binatang Ragunan
lon = 106.8180
point = ee.Geometry.Point([lon, lat])

start_date = '2015-01-01'
end_date = '2024-01-01'

# A. CHIRPS (Curah Hujan Satelit)
chirps = ee.ImageCollection("UCSB-CHG/CHIRPS/DAILY").filterDate(start_date, end_date)
data_raw = chirps.getRegion(point, scale=5500).getInfo()
header = data_raw[0]
df_chirps = pd.DataFrame(data_raw[1:], columns=header)
df_chirps['date'] = pd.to_datetime(df_chirps['time'], unit='ms')
df_chirps = df_chirps.set_index('date').sort_index()
df_chirps['Hujan'] = pd.to_numeric(df_chirps['precipitation']).fillna(0)

# B. NASA POWER (Suhu, Lembab, Angin, Tekanan)
str_start = start_date.replace('-', '')
str_end = end_date.replace('-', '')
base_url = "https://power.larc.nasa.gov/api/temporal/daily/point"
params = {
    'parameters': 'T2M,RH2M,WS2M,PS',
    'community': 'RE',
    'longitude': lon,
    'latitude': lat,
    'start': str_start,
    'end': str_end,
    'format': 'JSON'
}

try:
    response = requests.get(base_url, params=params, timeout=30)
    data_nasa = response.json()
    feat = data_nasa['properties']['parameter']
    df_nasa = pd.DataFrame([feat['T2M'], feat['RH2M'], feat['WS2M'], feat['PS']]).T
    df_nasa.columns = ['Suhu', 'Kelembaban', 'Angin', 'Tekanan']
    df_nasa.index = pd.to_datetime(df_nasa.index, format='%Y%m%d')
    df_nasa = df_nasa.replace(-999, np.nan).fillna(method='ffill')
    print("   âœ… Data Downloaded Sukses.")
except Exception as e:
    print(f"   âŒ Gagal Download NASA: {e}")
    raise e

# =============================================================================
# 2. FEATURE ENGINEERING
# =============================================================================
print("ðŸš€ [2/5] Feature Engineering...")

df = pd.concat([df_chirps[['Hujan']], df_nasa], axis=1, join='inner')

# Lag Features (Masa Lalu)
df['Lag1'] = df['Hujan'].shift(1)
df['Lag2'] = df['Hujan'].shift(2)
# Mean 7 Hari Mundur (Shift 1 biar tidak bocor)
df['Mean_7d'] = df['Hujan'].shift(1).rolling(7).mean()

# Fitur Interaksi (Indeks Kelembaban * Suhu)
df['Thermo'] = df['Suhu'] * df['Kelembaban']

# Fitur Waktu (Musiman)
day = 24 * 60 * 60
year = (365.2425) * day
seconds = df.index.map(pd.Timestamp.timestamp)
df['Sin_Time'] = np.sin(seconds * (2 * np.pi / year))
df['Cos_Time'] = np.cos(seconds * (2 * np.pi / year))

df = df.dropna()

features = ['Suhu', 'Kelembaban', 'Angin', 'Tekanan', 'Lag1', 'Lag2', 'Mean_7d', 'Thermo', 'Sin_Time', 'Cos_Time']
target = ['Hujan']

# Scaling
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X = scaler_X.fit_transform(df[features])
y = scaler_y.fit_transform(df[target])

# Windowing (Sequence 14 hari)
LOOK_BACK = 14
def create_dataset(X, y, look_back):
    Xs, ys = [], []
    for i in range(len(X) - look_back):
        Xs.append(X[i:(i + look_back)])
        ys.append(y[i + look_back])
    return np.array(Xs), np.array(ys)

X_seq, y_seq = create_dataset(X, y, LOOK_BACK)

# Split 80:20
split = int(len(X_seq) * 0.8)
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]

# =============================================================================
# 3. MODEL: BIDIRECTIONAL LSTM + HUBER LOSS
# =============================================================================
print("ðŸš€ [3/5] Training Robust Regression Model...")

# Kita menggunakan Bidirectional LSTM untuk menangkap pola hujan dari dua arah waktu
#

model = Sequential()

# Bidirectional: Membaca data dari kiri-kanan dan kanan-kiri
model.add(Bidirectional(LSTM(64, return_sequences=False), input_shape=(LOOK_BACK, len(features))))
model.add(Dropout(0.3))

# Dense Layer
model.add(Dense(32, activation='relu'))

# Output Layer pakai ReLU: Memaksa hasil >= 0 (Hujan ga mungkin negatif)
model.add(Dense(1, activation='relu'))

# OPTIMIZER & LOSS
# Huber Loss digunakan agar model tidak rusak saat menemui data banjir ekstrem (outlier)
opt = Adam(learning_rate=0.0005)
model.compile(optimizer=opt, loss=tf.keras.losses.Huber(delta=1.0), metrics=['mae'])

# Callbacks
es = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5)

history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[es, lr],
    verbose=1
)

# =============================================================================
# 4. EVALUASI
# =============================================================================
print("ðŸš€ [4/5] Evaluasi...")

preds = scaler_y.inverse_transform(model.predict(X_test))
actual = scaler_y.inverse_transform(y_test)

# Pastikan 0 tetap 0
preds = np.maximum(preds, 0)

rmse = np.sqrt(mean_squared_error(actual, preds))
mae = mean_absolute_error(actual, preds)
r2 = r2_score(actual, preds)

print(f"\n========================================")
print(f"   HASIL AKHIR (LOKASI: RAGUNAN)")
print(f"========================================")
print(f"   RMSE     : {rmse:.2f} mm")
print(f"   MAE      : {mae:.2f} mm")
print(f"   R2 SCORE : {r2:.4f}")
print(f"========================================")

# =============================================================================
# 5. VISUALISASI & VALIDASI VISUAL
# =============================================================================
# Kita ambil momen dimana hujan Aktual > 10mm (Hujan Sedang-Lebat)
idx_big_rain = np.where(actual > 10)[0]

if len(idx_big_rain) > 0:
    limit = min(50, len(idx_big_rain))
    subset_idx = idx_big_rain[:limit]

    plt.figure(figsize=(12, 6))
    plt.plot(actual[subset_idx], label='Aktual (Hujan > 10mm)', marker='o', color='blue')
    plt.plot(preds[subset_idx], label='Prediksi Model', marker='x', color='red', linestyle='--')
    plt.title(f"Performa Model Saat Hujan Deras (Ragunan)\nR2 Overall: {r2:.4f}")
    plt.ylabel("Curah Hujan (mm)")
    plt.legend()
    plt.show()
else:
    print("Tidak ada hujan > 10mm di data test untuk di-plot khusus.")

# Scatter Plot
plt.figure(figsize=(6,6))
plt.scatter(actual, preds, alpha=0.3)
plt.plot([0, actual.max()], [0, actual.max()], 'k--', color='red')
plt.xlabel("Aktual")
plt.ylabel("Prediksi")
plt.title("Scatter Plot: Sebaran Prediksi")
plt.show()

# ALERT
besok = preds[-1][0]
print(f"\nPrediksi Hujan Besok di Ragunan: {besok:.2f} mm")
```

```python
import ee
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

# Kunci agar hasil konsisten setiap di-run
tf.random.set_seed(42)
np.random.seed(42)

# ==========================================
# 1. KONEKSI & AMBIL DATA (CEPAT & AMAN)
# ==========================================
MY_PROJECT = 'plenary-vim-481714-s8' # Jakarta Selatan

print("1. Menghubungkan ke GEE...")
try:
    ee.Initialize(project=MY_PROJECT)
except:
    ee.Authenticate()
    ee.Initialize(project=MY_PROJECT)

print("2. Download Data CHIRPS...")
point = ee.Geometry.Point([106.8106, -6.2615])

# Ambil data 10 tahun terakhir
chirps = ee.ImageCollection("UCSB-CHG/CHIRPS/DAILY") \
    .filterDate('2014-01-01', '2024-01-01')

# Menggunakan getRegion (Jauh lebih cepat & stabil dibanding map loop)
try:
    data_raw = chirps.getRegion(point, scale=5500).getInfo()
    print(f"   \u2705 Data berhasil ditarik! ({len(data_raw)} baris)")
except Exception as e:
    print(f"   \u274c Gagal ambil data. Cek Project ID/Koneksi. Error: {e}")
    raise e

# ==========================================
# 2. DATA PREPARATION
# ==========================================
# Convert list to DataFrame
header = data_raw[0]
df = pd.DataFrame(data_raw[1:], columns=header)
df['date'] = pd.to_datetime(df['time'], unit='ms')
df = df.set_index('date').sort_index()
df['Hujan'] = pd.to_numeric(df['precipitation']).fillna(0)

# Define TIME_STEPS for 1 week
TIME_STEPS = 7

# Prepare data for sequence creation (just the 'Hujan' column)
data_to_sequence = df['Hujan'].values.reshape(-1, 1)

# SCALING (Wajib untuk LSTM)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data_to_sequence)

# Function to create sequences (time steps)
def create_dataset(dataset, time_steps=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_steps):
        a = dataset[i:(i + time_steps), 0] # Use 'time_steps' values as features
        dataX.append(a)
        dataY.append(dataset[i + time_steps, 0]) # Predict the next value
    return np.array(dataX), np.array(dataY)

# Create sequences
X_final, y_final = create_dataset(scaled_data, TIME_STEPS)

# RESHAPE UNTUK LSTM [Samples, Time Steps, Features]
# Features = 1 (only 'Hujan' value per time step)
X_final = np.reshape(X_final, (X_final.shape[0], X_final.shape[1], 1))
y_final = y_final.reshape(-1, 1) # Ensure y_final is 2D for inverse transform

# SPLIT DATA
split = int(len(X_final) * 0.8)
X_train, X_test = X_final[:split], X_final[split:]
y_train, y_test = y_final[:split], y_final[split:]

# ==========================================
# 3. MODEL: SIMPLE VANILLA LSTM
# ==========================================
print("3. Training Model...")

model = Sequential()

# --- INI INTI REKOMENDASINYA ---
# 1 Layer LSTM. Units=64 (Cukup pintar untuk tangkap pola).
# input_shape=(TIME_STEPS, 1) artinya 7 time steps, 1 feature.
model.add(LSTM(units=64, activation='tanh', input_shape=(TIME_STEPS, 1)))

# Output Layer (Wajib Ada Dense 1 untuk Regresi)
model.add(Dense(units=1, activation='linear'))
# -------------------------------

# Optimizer Adam dengan Learning Rate kecil agar belajar pelan tapi pasti
opt = Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss='mse')

# Train
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16, # Batch kecil update bobot lebih sering = lebih akurat
    validation_data=(X_test, y_test),
    verbose=1
)

# ==========================================
# 4. EVALUASI
# ==========================================
preds_scaled = model.predict(X_test)
preds_real = scaler.inverse_transform(preds_scaled)
y_test_real = scaler.inverse_transform(y_test)

# Koreksi minus (Hujan tidak mungkin negatif)
preds_real[preds_real < 0] = 0

rmse = np.sqrt(mean_squared_error(y_test_real, preds_real))
r2 = r2_score(y_test_real, preds_real)

print(f"\n=== HASIL AKHIR ===")
print(f"RMSE: {rmse:.2f} mm")
print(f"R2 Score: {r2:.4f}")

# Plotting
plt.figure(figsize=(12, 5))
# Plot 50 hari terakhir saja biar jelas
plt.plot(y_test_real[-50:], label='Hujan Asli', marker='.')
plt.plot(preds_real[-50:], label='Prediksi LSTM', linestyle='--')
plt.title(f'Prediksi Curah Hujan (Simple LSTM) - RMSE: {rmse:.2f}')
plt.ylabel('Curah Hujan (mm)')
plt.legend()
plt.show()
```
